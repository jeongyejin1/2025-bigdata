```python
import folium
anMap = folium.Map(location=[37.4489,127.1682],
               zoom_start=11,
               zoom_control=False,
               control_scale=True)
anMap
```

```python
from folium.plugins import MarkerCluster
```

```python
for n, row in uni.iterrows():
  mk=folium.Marker(location=[row['LA'], row['LO']], popup= folium.Popup(row['TITLE'], max_width=200))
  mk.add_to(anMap)
anMap
```

```python

mc=MarkerCluster().add_to(anMap)

for n, row in uni.iterrows():
  folium.Marker(location=[row['LA'], row['LO']],
                popup = folium.Popup(row['TITLE'], max_width=200)
                ).add_to(mc)
anMap
```
