```python
import pandas as pd
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/자연휴양림출력화일.xlsx')
df.head()
```

```python
import folium
```

```python
from folium.plugins import MarkerCluster
```

```python
anMap = folium.Map(location=[37.4489,127.1682],
               zoom_start=11,
               zoom_control=False,
               control_scale=True)
anMap
```

```python
for n, row in df.iterrows():
  print(n, row['휴양림명'], row['위도'], row['경도'])
```

```python
for n, row in df.iterrows():
  mk=folium.Marker(location=[row['위도'], row['경도']], popup= folium.Popup(row['휴양림명'], max_width=200))
  mk.add_to(anMap)
anMap
```

```python
anMap2 = folium.Map(location=[37.4489,127.1682],)
```

```python

mc=MarkerCluster().add_to(anMap2)

for n, row in df.iterrows():
  folium.Marker(location=[row['위도'], row['경도']],
                popup = folium.Popup(row['휴양림명'], max_width=200)
                ).add_to(mc)
anMap2
```
